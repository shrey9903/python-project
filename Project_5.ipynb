```python
#Name: Shrey Patel
#GR ID: 12376


print("RWPython OOP Project : Employee Management System")


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

class Employee(Person):
    def __init__(self, name, age, emp_id, salary):
        super().__init__(name, age)
        self.emp_id = emp_id
        self.__salary = salary

    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, value):
        if value > 0:
            self.__salary = value
    def display(self):
        super().display()
        print(f"Employee ID: {self.emp_id}")
        print(f"Salary: {self.__salary}")

class Manager(Employee):
    def __init__(self, name, age, emp_id, salary, department):
        super().__init__(name, age, emp_id, salary)
        self.department = department

    def display(self):
        super().display()
        print(f"Department: {self.department}")



person_1 = []
employee_1 = []
manager_1 = []

while True:
     print(" ")
     print ("Select an option:")
     print(" ")
     print("1. Create a Person")
     print("2. Create an Employe")
     print("3. Create a Manager")
     print("4. Show details")
     print("5. Exit")
     print(" ")

     choice = int(input("Enter your choice : "))
     print(" ")


     match choice:

        case 1:
            print("You have choice option 1: Create a Person")

            name = input("Enter Name: ")
            age = int(input("Enter Age: "))

            per = Person (name, age)
            person_1.append(per)

            print(f"\nPerson created with name: {name} and age: {age}\n")

        case 2:
            print("You have choice option 2: Create an Employee")

            name = input("Enter Name: ")
            age = int(input("Enter Age: "))
            emp_id = input("Enter Employee ID: ")
            salary = float(input("Enter Salary: "))

            emp = Employee (name, age, emp_id, salary)
            employee_1.append(emp)

            print(f"\nEmployee created with name: {name}, age: {age}, ID: {emp_id} and salary: {salary}\n")

        case 3:
            print("You have choice option 3: Create a Manager")

            name = input("Enter Name: ")
            age = int(input("Enter Age: "))
            emp_id = input("Enter Employee ID: ")
            salary = float(input("Enter Salary: "))
            dept = input("Enter Department: ")

            man = Manager (name, age, emp_id, salary, dept)
            manager_1.append(man)

            print(f"Manager created with name: {name}, age: {age}, ID: {emp_id}, salary: {salary} and department: {dept}.\n")

        case 4:
            print("You have choice option 4: Show details")
            print(" ")

            print("1. Show All Persons")
            print("2. Show All Employees")
            print("3. Show All Managers")
            print(" ")

            sub = int(input("Enter your choice: "))

            if sub == 1:
                if not person_1:
                    print("No persons are created")
                else:
                    for p in person_1:
                        p.display()
                        print(" ")

            elif sub == 2:
                if not employee_1:
                    print("No employees are created")
                else:
                    for e in employee_1:
                        e.display()
                        print(" ")

            elif sub == 3:
                if not manager_1:
                    print("No managers are created")
                else:
                    for m in manager_1:
                        m.display()
            else:
                print("Invalid choice")

        case 5:
            print ("Exiting the program. Goodbye!")
            break
```
