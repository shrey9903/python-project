```python
print("Welcome to the Student Data Organizer!")

l = {}

while True:
    print("""
Select an option:
1. Add Student
2. Display All Students
3. Update Student Information
4. Delete Student
5. Display Subjects Offered
6. Exit
""")

    x = int(input("Enter your choice: "))
    if x == 1:
        print("Enter the student details:")
        id = int(input("Student Id: "))
        name = input("Name: ")
        age = int(input("Age: "))
        grade = input("Grade: ")
        dob = input("Date of Birth (YYYY-MM-DD): ")
        sub = input("Subjects (comma separated): ")

        l[id] = {'id': id,'name': name,'age': age,'grade': grade,'dob': dob,'sub': sub}

        print("Student added successfully!")
    elif x == 2:
        print("Display All Students:")
        for i in l.values():
            print("Student ID:", i['id'],"| Name:", i['name'],"| Age:", i['age'],"| Grade:", i['grade'],"| Subjects:", i['sub'])
    elif x == 3:
        sid = int(input("Enter Student ID to update: "))
        if sid in l:
            l[sid]['name'] = input("New Name: ")
            l[sid]['age'] = int(input("New Age: "))
            l[sid]['grade'] = input("New Grade: ")
            l[sid]['dob'] = input("New DOB: ")
            l[sid]['sub'] = input("New Subjects: ")
            print("Student updated successfully!")
        else:
            print("Student not found.")
    elif x == 4:
        sid = int(input("Enter Student ID to delete: "))
        if sid in l:
            del l[sid]
            print("Student deleted successfully!")
        else:
            print("Student not found.")
    elif x == 5:
        print("Display Subjects Offered:")
        subjects = set()
        for i in l.values():
            for s in i['sub'].split(","):
                subjects.add(s.strip())
        for s in subjects:
            print("-", s)
    elif x == 6:
        print("Thanks for visiting the Student Data Organizer")
        break
```
