```python
# PR6 OOP File Operator
# Name Shrey Patel
# GR ID : 12376
# Batch : RW6
# Date : 03/12/25
import time

def write_entry(filename, entry):
    timestamp = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())
    with open (filename, "a") as file:
        file.write(f"{timestamp} - {entry}\n")

def read_entries(filename):
    try:
        with open (filename, "r") as file:
            return file.readlines()
    except FileNotFoundError:
        return []

def search_entry(filename, keyword):
    entries = read_entries(filename)
    matching_entries = [entry for entry in entries if keyword.lower() in entry.lower()]
    return matching_entries

def delete_all_entries(filename):
    try:
        with open(filename, "w") as file:
            file.write("")
        return True
    except FileNotFoundError:
        return False


print("Welcome to Journal Manager \n")

data = []
while True:

    print("Select an Option")
    print("01. Add a New Entry")
    print("02. View All Entries")
    print("03. Search Entries")
    print("04. Delete an Entry")
    print("05. Exit\n")

    choice = int(input("Enter your choice: "))

    match choice:

        case 1:
            entry = input("Enter your journal entry: ")
            write_entry("journal.txt", entry)
            print("Entry added successfully!\n")
            time.sleep(1)

        case 2:
            entries = read_entries("journal.txt")
            if entries:
                print("Loading entry...\n")
                for entry in entries:
                    print(entry.strip())
                    print(" ")
                    time.sleep(0.5)
            else:
                time.sleep(1)
                print("No Journal entries found.Start by adding a new entry.\n")

        case 3:
            print(" ")
            keyword = input("Enter a keyword or date to search: ")
            matching_entries = search_entry("journal.txt", keyword)
            if matching_entries:
                print(" ")
                time.sleep(1)
                print("Matching Entries:")
                time.sleep(0.5)
                for entry in matching_entries:
                    print(entry.strip())
                    print(" ")
                    time.sleep(0.5)
                print(" ")
            else:
                print(f"No entries were found for keyword: {keyword}\n")

        case 4:
            confirm = input("Are you sure you want to delete all entries? (yes/no): ")
            if confirm.lower() == "yes":
                if delete_all_entries("journal.txt"):
                    print(" ")
                    print("All journal entries have been deleted!\n")
                    print(" ")
                else:
                    print("No journal entries found to delete.")
            else:
                print("Deletion cancelled.")

        case 5:
            print("\nThankyou for using Personal Journal Manager. Goodbye!\n")
            break

        case _:
            print("Invalid choice. Please try again.\n")
```
